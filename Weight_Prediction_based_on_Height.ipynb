***Weight prediction based on Height which is given as input to the model.***

# ***Importing required libraries and Dataset.***

```
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

```
file = files.upload()
```

```
import io

dataset = pd.read_csv(io.StringIO(file['height-weight.csv'].decode('utf-8')))
dataset = dataset.drop(dataset.columns[[0,1,2]] , axis = 1)
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print(dataset.head())
```

**Visualise the data.**

```
plt.scatter(x , y)
plt.title('weight prediction based on height')
plt.xlabel('height')
plt.ylabel('weight')
plt.savefig('dataset.png')
plt.show()
```

***Split dataset into test and train sets.***

```
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state = 0)
```

# ***Build the model and train it.***

```
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg = reg.fit(x_train , y_train)
```

***Now the model training is done.***

# ***Make Predictions using the model.***

```
predictions = reg.predict(x_test)

print(predictions)
```

**Visualise the model.**

```
plt.plot(x_test , predictions)
plt.title('weight prediction based on height')
plt.xlabel('height')
plt.ylabel('weight')
plt.savefig('model.png')
plt.show()
```

# ***Evaluating the model.***

```
import math

rmse = math.sqrt(((predictions - y_test)**2).mean())
print(rmse)
```

```
plt.scatter(x_test , y_test , color = "blue")
plt.plot(x_test , predictions , color = "red")
plt.title('Weight prediction based on Height(Test Data)')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

**Random prediction.**

```
predicted_weight = reg.predict(150)
print("Predicted weight for height 150(cm) : %f" %(predicted_weight))
```
